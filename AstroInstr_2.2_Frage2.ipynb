```python
import matplotlib.pyplot as plt

b=2.898E-3

T= range(1,300)
L= [ ( b / t ) * 1E6 for t in T ]
plt.plot( T, L )
plt.title( "Strahlung eines schwarzen Körpers" )
plt.xlabel( "Temperatur [K]" )
plt.ylabel( "Wellenlänge λ [µm]" )
#plt.xscale('log')
plt.yscale('log')
plt.grid( True, linestyle="--" )
plt.show()
```
