```python
import pandas as pd
import matplotlib.pyplot as plt
import math

k= 1.380649E-23
G= 6.6743015E-11
mH= 1.6735575E-27
n= 1E11
Msol=1.988416E30

plt.subplot(1, 2, 1)
T= range(300)
M= [ ( ( ( 5.0 * k * t ) / ( G * mH ) )**(3.0/2.0) * ( 3 / ( 4.0 * math.pi * n * mH ) )**(1.0/2.0) ) / Msol for t in T ]
plt.plot( T, M, label="Mj", color="red" )
plt.fill_between( T, M, color="green", alpha=0.1)
plt.title( "Jeans Mass vs. Temperature" )
plt.xlabel( "Temperature [K]" )
plt.ylabel( r"Mass [M${\odot}$]" )
plt.grid( True, linestyle="--" )

plt.subplot(1, 2, 2)
T= 30
N= range(1,100) # E11
M= [ ( ( ( 5.0 * k * T ) / ( G * mH ) )**(3.0/2.0) * ( 3 / ( 4.0 * math.pi * n*1E11 * mH ) )**(1.0/2.0) ) / Msol for n in N ]
plt.plot( N, M, label="Mj", color="red" )
plt.fill_between( N, M, color="green", alpha=0.1)
plt.title( "Jeans Mass vs. Number Density" )
plt.xlabel( "Number Density [n10]" )
plt.ylabel( r"Mass [M${\odot}$]" )
plt.grid( True, linestyle="--" )

plt.tight_layout()
plt.show()
```
